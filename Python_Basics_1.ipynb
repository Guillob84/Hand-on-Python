# **Variables and data types**

```python
# --- Integer Data Type ---
# Integers are whole numbers, positive or negative, without decimals.
age = 40
print(f"Age: {age}, Type: {type(age)}")
```

```python
# --- Float Data Type ---
# Floats are numbers, positive or negative, containing one or more decimals.
price = 29.99
print(f"Price: {price}, Type: {type(price)}")
```

```python
# --- String Data Type ---
# Strings are sequences of characters, enclosed in single or double quotes.
name = "Alice"
message = 'Hello World'
print(f"Name: {name}, Type: {type(name)}")
print(f"Message: {message}, Type: {type(message)}")
```

```python
# --- Boolean Data Type ---
# Booleans represent one of two values: True or False.
is_active = True
is_admin = False
print(f"Is Active: {is_active}, Type: {type(is_active)}")
print(f"Is Admin: {is_admin}, Type: {type(is_admin)}")
```

```python
# --- List Data Type ---
# Lists are ordered, mutable (changeable) collections of items.
# They allow duplicate members and are enclosed in square brackets.
fruits = ["apple", "banana", "cherry"]
mixed_list = [1, "text", True, 3.14]
print(f"Fruits: {fruits}, Type: {type(fruits)}")
print(f"Mixed List: {mixed_list}, Type: {type(mixed_list)}")
```

```python
# --- Tuple Data Type ---
# Tuples are ordered, immutable (unchangeable) collections of items.
# They allow duplicate members and are enclosed in parentheses.
coordinates = (10, 20)
colors = ("red", "green", "blue")
print(f"Coordinates: {coordinates}, Type: {type(coordinates)}")
print(f"Colors: {colors}, Type: {type(colors)}")
```

```python
# --- Set Data Type ---
# Sets are unordered, unindexed, mutable collections of unique items.
# They do not allow duplicate members and are enclosed in curly braces.
unique_numbers = {1, 2, 3, 2, 1} # Duplicates are automatically removed
letters = {'a', 'b', 'c'}
print(f"Unique Numbers: {unique_numbers}, Type: {type(unique_numbers)}")
print(f"Letters: {letters}, Type: {type(letters)}")
```

```python
# --- Dictionary Data Type ---
# Dictionaries are unordered, mutable collections of key-value pairs.
# Keys must be unique and immutable, values can be of any type.
# They are enclosed in curly braces.
person = {
    "name": "Bob",
    "age": 25,
    "city": "New York"
}

student = {
    "id": 101,
    "courses": ["Math", "Science"],
    "is_enrolled": True
}
print(f"Person: {person}, Type: {type(person)}")
print(f"Student: {student}, Type: {type(student)}")
```

# **Loops (for / while)**

## For Loop Examples

```python
# 1. Iterating over a list
print("\n1. Iterating over a list:")
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(f"I like {fruit}")
```

```python
# 2. Iterating over a string
print("\n2. Iterating over a string:")
word = "Python"
for char in word:
    print(char)
```

```python
# 3. Iterating using range()
print("\n3. Iterating using range():")
# range(stop) - generates numbers from 0 up to (but not including) stop
print("Numbers from 0 to 4:")
for i in range(5):
    print(i)
```

```python
# range(start, stop) - generates numbers from start up to (but not including) stop
print("Numbers from 2 to 5:")
for i in range(2, 6):
    print(i)
```

```python

# range(start, stop, step) - generates numbers from start up to (but not including) stop, with a specified step
print("Even numbers from 0 to 10:")
for i in range(0, 11, 2):
    print(i)
```

```python
# range(start, stop, step) - generates numbers from 1 to 10, sharing the odd numbers
print("Odd numbers from 0 to 10:")
for i in range(1, 11, 2):
    print(i)
```

```python
# 4. Iterating with index using enumerate()
print("\n4. Iterating with index using enumerate():")
colors = ["red", "green", "blue"]
for index, color in enumerate(colors):
    print(f"Color at index {index}: {color}")
```

```python
# 5. Nested for loops
print("\n5. Nested for loops:")
for i in range(3):
    for j in range(2):
        print(f"({i}, {j})")
```

## While Loop Examples

```python
# 1. Basic while loop
print("\n1. Basic while loop (countdown):")
count = 5
while count > 0:
    print(count)
    count -= 1 # Decrement count in each iteration
print("Blast off!")
```

```python
# 2. While loop with a condition and break
print("\n2. While loop with a condition and break (guessing game):")
secret_number = 7
guess = 0
while True:
    guess += 1
    if guess == secret_number:
        print(f"You guessed the secret number {secret_number}!")
        break # Exit the loop when the condition is met
    elif guess > 10:
        print("Too many guesses, exiting.")
        break
    else:
        print(f"Attempt {guess}: Not {secret_number}")
```

```python
# 3. While loop with continue
print("\n3. While loop with continue (skipping even numbers):")
i = 0
while i < 10:
    i += 1
    if i % 2 == 0: # If i is even
        continue # Skip the rest of the current iteration and go to the next
    print(f"{i} is odd")
```

```python
# 5. While loop with else block (executed if loop finishes normally)
print("\n5. While loop with else block:")
i = 1
while i < 5:
    print(i)
    i += 1
else:
    print("Loop finished normally, i is now 5 or greater.")


print("\nLoops demonstration complete.")
```

# **Functions**

```python
# 1. Basic Function Definition
# A simple function that prints a greeting.
def greet():
    print("Hello, World!")
# Calling the function
greet()
```

```python
# 2. Function with Arguments
# Functions can take arguments (inputs).
def greet_name(name):
    print(f"Hello, {name}!")

greet_name("Alice")
greet_name("Bob")
```

```python
# 3. Function with Return Value
# Functions can return values using the 'return' keyword.
def add(a, b):
    return a + b

result = add(5, 3)
print(f"5 + 3 = {result}")
```

```python
# 4. Function with Default Arguments
# You can provide default values for arguments.
def greet_language(name, language="English"):
    if language == "English":
        print(f"Hello, {name}!")
    elif language == "Spanish":
        print(f"Hola, {name}!")
    else:
        print(f"Hi, {name}! (in {language})")

greet_language("Charlie") # Uses default language
greet_language("David", "Spanish") # Overrides default
```

```python
# 5. Functions with Arbitrary Arguments (*args and **kwargs)
# *args allows a function to accept a variable number of positional arguments.
def sum_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(f"Sum of 1, 2, 3: {sum_all(1, 2, 3)}")
print(f"Sum of 10, 20, 30, 40: {sum_all(10, 20, 30, 40)}")

# **kwargs allows a function to accept a variable number of keyword arguments.
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Eve", age=30, city="London")
```

```python
# 6. Lambda Functions (Anonymous Functions)
# Small, anonymous functions defined with the 'lambda' keyword.
# They can take any number of arguments, but can only have one expression.
multiply = lambda x, y: x * y
print(f"Lambda multiply(4, 5): {multiply(4, 5)}")
```

```python
# Using lambda with higher-order functions (like map, filter)
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x**2, numbers))
print(f"Squared numbers (using map and lambda): {squared_numbers}")
```

```python
# 7. Docstrings
# Explanations for functions, enclosed in triple quotes.
def calculate_area(length, width):
    """
    Calculates the area of a rectangle.

    Args:
        length (float): The length of the rectangle.
        width (float): The width of the rectangle.

    Returns:
        float: The area of the rectangle.
    """
    return length * width

print(f"Area of 10x5 rectangle: {calculate_area(10, 5)}")
print(f"Docstring for calculate_area: {calculate_area.__doc__}")
```

```python
# 8. Function Scope (Local vs. Global Variables)
# Variables defined inside a function are local to that function.
global_var = "I am global"

def scope_example():
    local_var = "I am local"
    print(f"Inside function: {local_var} and {global_var}")
    # global_var = "Changing global" # This would create a new local variable unless 'global' keyword is used

scope_example()
print(f"Outside function: {global_var}")
# print(local_var) # This would cause an error as local_var is not defined outside the function
```

```python
# 9. Recursion
# A function calling itself to solve a problem.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(f"Factorial of 5: {factorial(5)}") # 5*4*3*2*1 = 120
print(f"Factorial of 0: {factorial(0)}")
```

# **Basic data structures (lists, tuples, sets, dictionaries)**

## Lists in Data Analysis

Lists are excellent for storing ordered collections of data, like a series of measurements or records.

```python
# Data example
sales_data = [150, 200, 120, 300, 180, 250]
print(f"Original Sales Data: {sales_data}")
```

## Sets in Data Analysis

Sets are useful for storing unique items and performing set operations (union, intersection, difference).


```python
# List of categories from customer feedback (may contain duplicates)
feedback_categories = ["Service", "Product Quality", "Delivery", "Service",
                       "Pricing", "Product Quality"]
```

```python
# Find unique categories
unique_categories = set(feedback_categories)
print(f"Unique Feedback Categories: {unique_categories}")
```

```python
# Users who interacted with Feature A and Feature B
users_feature_A = {"Alice", "Bob", "Charlie", "David"}
users_feature_B = {"Charlie", "Eve", "Frank", "Bob"}
```

```python
# Users who interacted with BOTH features (intersection)
common_users = users_feature_A.intersection(users_feature_B)
print(f"Users interacting with both A and B: {common_users}")
```

```python
# All unique users who interacted with EITHER feature (union)
all_users = users_feature_A.union(users_feature_B)
print(f"All unique users (A or B): {all_users}")
```

## Dictionaries in Data Analysis

Dictionaries store data as key-value pairs, ideal for structured records or lookup tables.


```python
# Storing sensor data for different cities
city_temperatures = {
    "London": {"avg": 12.5, "max": 25, "min": 5},
    "Paris": {"avg": 14.0, "max": 28, "min": 7},
    "Tokyo": {"avg": 15.5, "max": 30, "min": 10}
}

print(f"Temperature data for London: {city_temperatures['London']}")
print(f"Average temperature in Paris: {city_temperatures['Paris']['avg']}°C")
```

```python
# Counting occurrences of items (e.g., product sales by type)
product_orders = ["Laptop", "Mouse", "Keyboard", "Laptop", "Mouse", "Monitor", "Laptop"]
product_counts = {}
for product in product_orders:
    product_counts[product] = product_counts.get(product, 0) + 1

print(f"Product Order Counts: {product_counts}")
```

```python
# Customer data (ID as key)
customer_data = {
    "CUST001": {"name": "Alice", "age": 30, "city": "New York"},
    "CUST002": {"name": "Bob", "age": 24, "city": "London"}
}

print(f"Customer CUST001 details: {customer_data['CUST001']}")
```

## File I/O: Handling CSV Files (Structured Data)

```python
import csv

csv_file_name = "products.csv"

# Data to write to CSV
products = [
    ["ID", "Name", "Price", "Stock"],
    [1, "Laptop", 1200, 50],
    [2, "Mouse", 25, 200],
    [3, "Keyboard", 75, 150]
]
```

```python
# Writing to a CSV file
with open(csv_file_name, "w", newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerows(products)
print(f"Data written to {csv_file_name}")
```

```python
# Reading from a CSV file
print(f"\nContent of {csv_file_name}:")
with open(csv_file_name, "r") as csvfile:
    csv_reader = csv.reader(csvfile)
    header = next(csv_reader) # Read header row
    print(f"Header: {header}")
    print("Rows:")
    for row in csv_reader:
        print(row)
```

## More Advanced CSV Reading (using DictReader for easier access)

```python
print("\n--- Advanced CSV Reading (DictReader) ---")
with open(csv_file_name, "r") as csvfile:
    csv_dict_reader = csv.DictReader(csvfile)
    print("Rows (using DictReader):")
    for row in csv_dict_reader:
        print(f"Product Name: {row['Name']}, Price: ${row['Price']}")

print("\nInput/Output demonstrations complete.")
```

# **Basic error handling**

## Basic Error Handling: try-except


```python
# 1. Handling ValueError (e.g., invalid input conversion)
print("\n1. Handling ValueError:")
def get_numeric_input():
    while True:
        try:
            user_input = input("Enter a number: ")
            number = float(user_input)
            return number
        except ValueError:
            print("Invalid input. Please enter a valid number.")
# Uncomment the line below to test the function:
num = get_numeric_input()
print(f"You entered: {num}")
```

```python
# 2. Handling ZeroDivisionError
print("\n2. Handling ZeroDivisionError:")
def divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Error: Cannot divide by zero!")
        return None
    else:
        print(f"Division successful. Result: {result}")
        return result

divide(10, 2)
divide(10, 0)
```

```python
# 3. Handling FileNotFoundError
print("\n3. Handling FileNotFoundError:")
def read_file_content(filename):
    try:
        with open(filename, 'r') as file:
            content = file.read()
            print(f"Content of {filename}:\n{content[:50]}...") # Print first 50 chars
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

read_file_content("my_data.txt") # This file exists from previous I/O example
read_file_content("non_existent_file.txt")
```

```python
# 4. Using try-except-else-finally
print("\n4. Using try-except-else-finally:")
def process_data(data_list, index):
    try:
        value = data_list[index]
        if value == 0:
            raise ValueError("Value cannot be zero for further processing.")
        processed_value = 100 / value
    except IndexError:
        print("Error: Index out of bounds!")
        processed_value = None
    except ZeroDivisionError:
        print("Error: Attempted division by zero during processing!")
        processed_value = None
    except ValueError as ve:
        print(f"Data processing error: {ve}")
        processed_value = None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        processed_value = None
    else:
        print(f"Processing successful. Original value: {value}, Processed: {processed_value:.2f}")
    finally:
        print("Data processing attempt finished.")
    return processed_value

my_numbers = [10, 5, 0, 20]

process_data(my_numbers, 0) # Valid division
process_data(my_numbers, 2) # ValueError (explicitly raised)
process_data(my_numbers, 1) # Valid division
process_data(my_numbers, 4) # IndexError
```

## Advanced Error Handling: Custom Exceptions

```python
class DataValidationException(Exception):
    """Custom exception for data validation errors."""
    def __init__(self, message, invalid_data=None):
        super().__init__(message)
        self.invalid_data = invalid_data

def analyze_temperature_data(temperatures):
    if not isinstance(temperatures, list):
        raise TypeError("Input must be a list of temperatures.")
    if not temperatures:
        raise DataValidationException("Temperature list cannot be empty.")

    valid_temps = []
    for temp in temperatures:
        try:
            temp_float = float(temp)
            if not (-50 <= temp_float <= 50): # Assuming reasonable temperature range in Celsius
                raise DataValidationException(
                    f"Temperature {temp_float}°C out of expected range (-50 to 50).",
                    invalid_data=temp
                )
            valid_temps.append(temp_float)
        except ValueError:
            raise DataValidationException(
                f"Non-numeric temperature value found: '{temp}'.",
                invalid_data=temp
            )
    avg_temp = sum(valid_temps) / len(valid_temps)
    print(f"Average valid temperature: {avg_temp:.2f}°C")
    return avg_temp

print("\nTesting analyze_temperature_data:")
```

```python
# Example 1: Valid data
try:
    analyze_temperature_data([10, 15.5, 12])
except (TypeError, DataValidationException) as e:
    print(f"Caught error: {e}")
```

```python
# Example 2: Empty list
try:
    analyze_temperature_data([])
except (TypeError, DataValidationException) as e:
    print(f"Caught error: {e}")
```

```python
# Example 3: Invalid type
try:
    analyze_temperature_data("not a list")
except (TypeError, DataValidationException) as e:
    print(f"Caught error: {e}")
```

```python
# Example 4: Non-numeric data
try:
    analyze_temperature_data([20, 'abc', 25])
except (TypeError, DataValidationException) as e:
    print(f"Caught error: {e}. Invalid data: {e.invalid_data if hasattr(e, 'invalid_data') else 'N/A'}")
```

```python
# Example 5: Out of range data
try:
    analyze_temperature_data([5, 60, 10])
except (TypeError, DataValidationException) as e:
    print(f"Caught error: {e}. Invalid data: {e.invalid_data if hasattr(e, 'invalid_data') else 'N/A'}")
```
